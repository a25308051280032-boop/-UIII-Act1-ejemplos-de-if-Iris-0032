#### Ejemplos de if, if else, if elif

```python
# Ejemplo 1: Verificación de edad (Relacionales)
# Este código determina si una persona es mayor de edad basándose en una comparación simple
edad = int(input("Ingresa tu edad: ")) # Convierte la entrada del usuario a un entero [8]
if edad >= 18: # Utiliza el operador relacional 'mayor o igual que' [4]
    print("Eres mayor de edad.") # Se ejecuta si la condición es verdadera [1]
else: # Maneja el caso en que la condición del 'if' sea falsa [1]
    print("Eres menor de edad.") # Indica que la persona es menor de 18 años
```

```python
# Ejemplo 2: Clasificación de números y operaciones (Aritméticos y Relacionales)
# En este ejemplo se evalúa si un número es positivo, negativo o cero, utilizando una estructura múltiple

numero = int(input("Ingresa un número: ")) # Recibe y convierte el número [8]
if numero > 0: # Compara si el número es mayor a cero [4]
    print("El número es positivo.") # Resultado si la comparación es verdadera
elif numero < 0: # Evalúa una segunda condición si la primera falló [1]
    print("El número es negativo.") # Resultado si el número es menor a cero [4]
else: # Se ejecuta si no es ni mayor ni menor a cero [1]
    print("El número es cero.") # Resultado final lógico
```

```python
# Ejemplo 3: Cálculo de año bisiesto (Lógicos, Aritméticos y Relacionales)
# Este ejemplo es más complejo ya que combina cálculos matemáticos con lógica booleana para determinar una propiedad del calendario

anio = int(input("Ingresa un año: ")) # Solicita el año al usuario
# Se usa el operador '%' para el residuo de la división y operadores lógicos 'and'/'or' [3, 4]
if (anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0):
    print(f"{anio} es un año bisiesto.") # El año cumple las reglas matemáticas de bisiesto [4]
else: # Si la lógica compuesta del 'if' resulta falsa [1]
    print(f"{anio} no es un año bisiesto.") # El año no es bisiesto
```

```python
# Ejemplo 4: Descuento por volumen de compra (Lógicos y Relacionales)
# Aquí se muestra cómo verificar si un valor se encuentra dentro de un rango específico utilizando una sintaxis simplificada de Python

cantidad = int(input("¿Cuántos artículos compraste? ")) # Captura la cantidad de compra
if 10 <= cantidad <= 20: # Verifica si la cantidad está en el rango de 10 a 20 inclusive [9]
    print("Tienes un descuento especial por compra mediana.") # Mensaje si está en el rango
elif cantidad > 20: # Evalúa si la compra es mayor al rango previo [4]
    print("Tienes el descuento máximo.") # Aplicable a compras grandes
else: # Caso para cantidades menores a 10 [1]
    print("No aplica descuento especial.") # Mensaje por defecto
```
